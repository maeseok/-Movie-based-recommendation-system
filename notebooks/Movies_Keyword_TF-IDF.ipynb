```python
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
from gliner import GLiNER

okt = Okt() #명사 추출용

model = GLiNER.from_pretrained("urchade/gliner_multi")  # NER 처리용
labels = ["Person", "Age", "Place", "Organization", "Date"]

def ner(text):
    ner_results = model.predict_entities(text, labels)
    filtered_text = text

    for entity in ner_results:
        if entity["label"] == 'Person':
            filtered_text = filtered_text.replace(entity["text"], '')  # 사람으로 판명된 개체명은 " "으로 대체됨.
        if entity["label"] == 'Age':
            filtered_text = filtered_text.replace(entity["text"], '')
        if entity["label"] == 'Place':
            filtered_text = filtered_text.replace(entity["text"], '')
        if entity["label"] == 'Organization':
            filtered_text = filtered_text.replace(entity["text"], '')
        if entity["label"] == 'Date':
            filtered_text = filtered_text.replace(entity["text"], '')

    return filtered_text


def preprocess_text(text):
    results = [] #명사 토큰을 담을 리스트
    filtered_tokens = [] #명사 토큰 중 불용어를 처리하고 난 다음 전처리 토큰을 담을 리스트

    text = ner(text) #고유명사 제거

    for token, pos in okt.pos(text):
        if len(token) != 1 and pos == "Noun": #2글자 이상의 명사 키워드만 반환
            results.append(token)

    stopwords = ["그녀", "대한", "때문", "여러", "정도", "더욱", "오히려", "한편", "때로는", "다가" "모습", "대한"] #불용어 정의
    filtered_tokens = [token for token in results if token not in stopwords]

    return filtered_tokens

# 데이터 파일 불러오기 (인코딩 지정)
file_path = "/content/drive/MyDrive/new_movie_data.csv"
df = pd.read_csv(file_path, header=0, encoding='cp949')
df_cleaned = df.dropna()

# 8번째 열에서 문자열이 아닌 행 삭제
df_cleaned = df_cleaned[df_cleaned.iloc[:, 7].apply(lambda x: isinstance(x, str))]

# 빠른 모델 성능 체크를 위해 상위 7150개 데이터만 추출
df_cleaned7150 = df_cleaned.iloc[3500:7150]

# 8번째 열 전처리: 명사만 남기고 전처리
df_cleaned7150['processed_text'] = df_cleaned7150.iloc[:, 7].apply(preprocess_text)

#tf-idf를 위한 문자열 변환
df_cleaned7150['processed_text'] = df_cleaned7150['processed_text'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# TF-IDF 계산
vectorizer = TfidfVectorizer(max_df=0.85, max_features=10000, token_pattern=r'\b\w+\b')
tfidf_matrix = vectorizer.fit_transform(df_cleaned7150['processed_text'])

# TF-IDF로 키워드 추출 함수 정의
def extract_keywords_tfidf(tfidf_matrix, feature_names, top_n=10):
    keywords = []
    for i in range(tfidf_matrix.shape[0]):
        row = tfidf_matrix[i]
        indices = row.indices
        scores = row.data
        top_indices = indices[scores.argsort()[-top_n:]]  # 상위 n개의 인덱스
        top_keywords = [feature_names[index] for index in top_indices]
        keywords.append(", ".join(top_keywords))
    return keywords

# 키워드 추출
feature_names = vectorizer.get_feature_names_out()
df_cleaned7150['키워드'] = extract_keywords_tfidf(tfidf_matrix, feature_names)

# 결과 확인
df_cleaned7150
```

```python
#결과물 저장
num = 7150
output_file_path = f'/content/drive/MyDrive/movie_keyword{num}.csv'
df_cleaned7150.to_csv(output_file_path, index=False, encoding='utf-8-sig')
```
